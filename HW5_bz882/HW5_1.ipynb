```python
# put samples of distributions(don't need to be gaussion) with s and std--> generate a standard distribution(mean = 0, std = 1)
#based on Z
import numpy as np
import pylab as pl
import pandas as pd
import scipy.stats as st
%pylab inline
```

```python
poisson = st.poisson.rvs(mu=100,size=1000)
```

## 1. Plot a Poisson Distribution

```python
pl.hist(poisson)
pl.title('Poisson Distribution ($\lambda = 100$)', fontsize=18)
pl.xlabel('$\mu$',fontsize=18)
pl.ylabel('frequency', fontsize=18);
```

```python
def calculate_z(poiss, mu = 100, std = 10):
    z = np.sqrt(len(poiss)) * (mu - poiss.mean()) / std
    print (len(poiss), mu, poiss.mean(), std)
    return z
```

```python
calculate_z(poisson)
```

```python
# Generate 1000 samples of size 1000 by Poisson distribution
z = np.zeros(1000)
mymean = np.zeros(1000)
np.random.seed(999)
for i in range(1000):
    poiss = st.poisson.rvs(mu = 100, size = 1000)
    mymean[i] = poiss.mean()
    z[i] = (calculate_z(poiss))
```

## 2. plot at least one of the samples

```python
pl.hist(poiss)
pl.title('Poisson Distribution ($\lambda = 100$)', fontsize=18)
pl.xlabel('$\mu$',fontsize=18)
pl.ylabel('frequency', fontsize=18);
```

## 3. plot the distribution of z statistics

```python
pl.hist(z)
pl.title('Distribution of Z-score', fontsize=18)
pl.xlabel('z-statistics',fontsize=18)
pl.ylabel('frequency', fontsize=18);
```

## 4. plot the distribution of sample means

```python
pl.hist(mymean)
pl.title('Distribution of 1000 sample means', fontsize=18)
pl.xlabel('Sample means',fontsize=18)
pl.ylabel('frequency', fontsize=18);
```

## 5. Fit the z-statistics with a gaussian model and assess the goodness of fit with a simple test (AD or KS).

```python
st.kstest(z,'norm')
```

#### We set significance level as 0.05.
#### Null Hypothesis: The distribution of z-statistcs of 1000 sample means generated from poission fits the gaussian model.
#### Since the P-value of the ks test (0.5) is higher than our significance level 0.05, we cannot reject the null hypothesis.
#### We conclude that the distribution of Z statistics of sample means follows the Gaussian distribution.
