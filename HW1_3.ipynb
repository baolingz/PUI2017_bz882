That becomes possible because computers are unable to poduce truly random numbers. When we ask a computer to generate a random number, it will in fact generate a "pseudo random" number. The number will be randon in the sanse that it will be some number between, for example, 0 and 1 if we ask for a uniform random distribution or a number near 0 if we aks for a Gaussian distribution with mean 0. But the sequence of pseudo-random numbers that a compujter generates will depend on its **seed**: if you know the seed you know all the numbers that are "randomely" drawn.

You can read more about pseudorandom number generation in this wikipedia article https://en.wikipedia.org/wiki/Pseudorandom_number_generator
